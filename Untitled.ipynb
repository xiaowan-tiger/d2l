# 1. 线性回归的从零开始实现

```python
%matplotlib inline
import random
import torch
from d2l import torch as d2l

import numpy as np
from torch.utils import data
```

```python
def synthetic_data(w, b, num_examples):
    X = torch.normal(0, 1, (num_examples, len(w))) # normal(mean, std, shapeOflist)
    y = torch.matmul(X, w) + b
    y += torch.normal(0,0.01, y.shape)
    return X, y.reshape((-1, 1))
```

```python
true_w = torch.tensor([2, -3.4])
true_b = 4.2
features, labels = synthetic_data(true_w, true_b, 1000)
```

```python
d2l.set_figsize()
d2l.plt.scatter(features[:, 1].detach().numpy(), labels.detach().numpy(), 1)
```

```python
def data_iter(batch_size, features, labels):
    num_examples = len(labels)
    indices = list(range(num_examples))
    random.shuffle(indices)
    
    # 遍历整个数据集
    for i in range(0, num_examples,batch_size):
        # small batch data - indices
        batch_indices = torch.tensor(
            indices[i:min(i+batch_size, num_examples)])
        yield features[batch_indices], labels[batch_indices]
```

```python
def linear_reg(X, w, b):
    return torch.matmul(X,w) + b

def squared_loss(y_hat, y):
    return (y_hat - y.reshape(y_hat.shape))**2 / 2

def small_gradient_decent(params, lr, batch_size):
    with torch.no_grad():
        for param in params:
            param -= lr * param.grad / batch_size
            param.grad.zero_()
```

```python
# 必须使用torch.Tensor 定义 w 和 b,而不应该仅仅为其赋值
w = torch.normal(0, 1, size=(2,1), requires_grad=True)
b = torch.zeros(1, requires_grad=True)
batch_size = 10
lr = 0.03
num_epochs = 3
net = linear_reg
loss = squared_loss

for epoch in range(num_epochs):
    for X, y in data_iter(batch_size, features, labels):
        l = loss(net(X, w, b), y)
        l.sum().backward()
        small_gradient_decent([w, b], lr, batch_size)
    # 单次迭代的训练误差：
    with torch.no_grad():
        train_l = loss(net(features, w, b), labels)
        print(f'epoch {epoch + 1}, loss {float(train_l.mean()):f}')
```

# 2. 简洁实现

```python
true_w = torch.tensor([2, -3.4, 5])
true_b = 4.2

# d2l synthetic_data
features, labels = d2l.synthetic_data(true_w, true_b, 1000)
```

```python
def load_array(data_arrays, batch_size, is_train=True):  #@save
    """构造一个PyTorch数据迭代器"""
    dataset = data.TensorDataset(*data_arrays)
    return data.DataLoader(dataset, batch_size, shuffle=is_train)
```

```python
new_data_iter = load_array((features, labels), batch_size)
```

```python
from torch import nn

net = nn.Sequential(nn.Linear(3, 1)) # 3: features counts, 1: output counts
```

```python
# initialize parameters - weight & bias
net[0].weight.data.normal_(0, 0.01)
net[0].bias.data.fill_(0)
```

```python
# define loss & trainer 
loss = nn.MSELoss()
trainer = torch.optim.SGD(net.parameters(), lr=0.03)
```

```python
for epoch in range(num_epochs):
    for X, y in new_data_iter:
        l = loss(net(X), y)
        trainer.zero_grad()
        l.backward()
        trainer.step()
    l = loss(net(features),labels)
    print(f'epoch {epoch + 1}, loss {l:f}')
```
